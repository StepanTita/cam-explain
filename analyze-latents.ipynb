```python
import matplotlib.pyplot as plt

# Dataset
# distilbert-imdb
results = {
    4: {'acc': 0.6494, 'f1': 0.6171, 'precision': 0.7258, 'recall': 0.6494},
    8: {'acc': 0.7111, 'f1': 0.7053, 'precision': 0.7291, 'recall': 0.7111},
    16: {'acc': 0.7790, 'f1': 0.7784, 'precision': 0.7819, 'recall': 0.7790},
    32: {'acc': 0.7652, 'f1': 0.7650, 'precision': 0.7658, 'recall': 0.7652},
    64: {'acc': 0.7761, 'f1': 0.7760, 'precision': 0.7766, 'recall': 0.7761},
    128: {'acc': 0.7958, 'f1': 0.7957, 'precision': 0.7965, 'recall': 0.7958},
}

# Extracting the data
latents = list(results.keys())
acc = [results[k]['acc'] for k in latents]
f1 = [results[k]['f1'] for k in latents]
precision = [results[k]['precision'] for k in latents]
recall = [results[k]['recall'] for k in latents]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(latents, acc, marker='o', label='Accuracy')
plt.plot(latents, f1, marker='o', label='F1 Score')
plt.plot(latents, precision, marker='o', label='Precision')
plt.plot(latents, recall, marker='o', label='Recall')

# Adding titles and labels
plt.title('Metrics vs Latents')
plt.xlabel('Latents')
plt.ylabel('Metric Values')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
